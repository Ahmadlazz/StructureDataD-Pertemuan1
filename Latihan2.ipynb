Input 10 data dalam sebuah list

```python
list = []
for i in range(10):
    angka = int(input("Masukkan angka: "))
    list.append(angka)
```

Menampilkan nilai maksimum, minimum, dan rata-rata

```python
print("List:", list)
print("Nilai maksimum:", max(list))
print("Nilai minimum:", min(list))
print("Nilai rata-rata:", sum(list)/len(list))
```

membuat grafik batang menggunakan matplotlib

```python
import matplotlib.pyplot as plt
plt.bar(range(len(list)), list)
plt.title("Grafik Batang")
plt.xlabel("Index")
plt.ylabel("Nilai")
plt.show()
```
