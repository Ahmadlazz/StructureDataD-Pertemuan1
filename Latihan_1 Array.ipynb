5 Nama Mahasiswa


```
mahasiswa = ["Andi", "Budi", "Dewi", "Eko"]
print("List Mahasiswa: ")
print(mahasiswa)
```

Tambahkan 1 nama baru dengan append

```
mahasiswa.append("Agus")
print("Setelah ditambahkan nama baru")
print(mahasiswa)
```

Hapus satu elemen menggunakan remove

```
mahasiswa.remove("Budi")
print("Setelah menghapus nama Budi")
print(mahasiswa)
```
