Menginput 10 nilai

```python
import numpy as np
print("Memasukkan 10 Nilai:  ")
nilai = np.array([88, 90, 85, 77, 100, 95, 87, 55, 99, 72])
print(nilai)
```

Menampilkan Nilai Tertinggi dan Terendah dari 10 nilai

```python
print("Menampilkan nilai tertinggi dan terendah:  ")
print_tertinggi = nilai.max()
print(nilai)
print("Nilai Tertinggi: ", nilai.max())
print("Nilai Terendah: ", nilai.min())
```

Menghitung Rata-rata

```python
print("Menghitung nilai rata-rata:  ")
rata_rata = nilai.mean()
print("Array: ", nilai)
print("Rata-rata: ", nilai.mean())
```

Hitung jumlah mahasiswa lulus (>= 60)

```python
print("Hitung jumlah mahasiswa lulus (>= 60):  ")
lulus = nilai >= 60
jumlah_lulus = np.sum(lulus)
print("Jumlah mahasiswa yang lulus: ", jumlah_lulus)
```

Buat grafik untuk menggambarkan nilai tertinggi dan terendah

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.plot(nilai, marker='o', linestyle='-', color='blue', label='Semua Nilai')

max_val = np.max(nilai)
min_val = np.min(nilai)
max_idx = np.argmax(nilai)
min_idx = np.argmin(nilai)


plt.title('Grafik Nilai Mahasiswa dengan Titik Tertinggi dan Terendah')
plt.xlabel('Indeks Nilai')
plt.ylabel('Nilai')
plt.grid(True)
plt.legend()
plt.show()
```

Buat grafik untuk menggambarkan data kelulusan

```python
import matplotlib.pyplot as plt
import numpy as np

status_kelulusan = np.where(nilai >= 60, 'Lulus', 'Tidak Lulus')


status_counts = np.unique(status_kelulusan, return_counts=True)

labels = status_counts[0]
counts = status_counts[1]

plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color=['green', 'red'])

plt.title('Distribusi Kelulusan Mahasiswa')
plt.xlabel('Status Kelulusan')
plt.ylabel('Jumlah Mahasiswa')

for i, count in enumerate(counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Analisis kompleksitas tiap operasi

```python
print("Analisis Kompleksitas Operasi:  ")
print("Labels (Status Kelulusan):", labels)
print("Counts (Jumlah Mahasiswa):", counts)
```
